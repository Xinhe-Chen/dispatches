```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import pathlib
```

## NE case (10,500)

```python
# for original data
rev_data = pathlib.Path('NE_revenue.csv')
input_data = pathlib.Path('sweep_parameters_results_NE_whole.h5')
df_rev = pd.read_csv(rev_data)
df_input = pd.read_hdf(input_data)
rev = df_rev.iloc[0:48].to_numpy()[:,1]/1000000
```

```python
# for predicted data
rev_data = pathlib.Path('NE_predicted_revenue_new.csv')
input_data = pathlib.Path('sweep_parameters_results_NE_whole.h5')
df_rev_pred = pd.read_csv(rev_data)
df_input = pd.read_hdf(input_data)
rev_pred = df_rev_pred.iloc[0:48].to_numpy()[:,1]/1000000
```

```python
# calculate the mean error
err = []
for i, j in zip(rev,rev_pred):
    # relative error
    err.append((np.abs(i-j))/i*100)
print(f'The mean error is {np.round(np.mean(err),2)}%')
```

```python
# for original data
pem_ratio = np.array([0.05,0.1,0.15,0.2,0.25,0.3,0.4,0.5])
h2_price = np.array([15,20,25,30,35,40])/20
X,Y = np.meshgrid(pem_ratio,h2_price)
rev_contour = np.zeros((len(pem_ratio),len(h2_price)))
c = 0
for p in range(len(pem_ratio)):
    for h in range(len(h2_price)):
        rev_contour[p,h] = rev[c]
        c += 1
# np.shape(rev_contour)
contour = plt.contour(X,Y,rev_contour.T)
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=10)
# plt.scatter(X,Y)
plt.xlabel('Ratio of capacities of PEM and NPP')
plt.ylabel('Selling price of Hydrogen [$/kg]')
plt.title('Electricity Revenue (original) [Million USD] (10,500)')
```

```python
# for predicted data
pem_ratio = np.array([0.05,0.1,0.15,0.2,0.25,0.3,0.4,0.5])
h2_price = np.array([15,20,25,30,35,40])/20
X,Y = np.meshgrid(pem_ratio,h2_price)
rev_contour = np.zeros((len(pem_ratio),len(h2_price)))
c = 0
for p in range(len(pem_ratio)):
    for h in range(len(h2_price)):
        rev_contour[p,h] = rev_pred[c]
        c += 1
# np.shape(rev_contour)
contour = plt.contour(X,Y,rev_contour.T)
plt.clabel(contour, colors = 'k', fmt = '%2.1f', fontsize=10)
# plt.scatter(X,Y)
plt.xlabel('Ratio of capacities of PEM and NPP')
plt.ylabel('Selling price of Hydrogen [$/kg]')
plt.title('Electricity Revenue (predicted) [Million USD] (10,500)')
```

```python
# plot the revenue at pem ratio = 0.4
x = df_input[36:42]['PEM_indifference_point'].to_numpy()/20
rev_needed = df_rev.iloc[36:42].to_numpy()[:,1]/1000000
rev_pred_needed = df_rev_pred.iloc[36:42].to_numpy()[:,1]/1000000
plt.plot(x, rev_needed, label = 'original')
plt.plot(x, rev_pred_needed, label = 'surrogate')
plt.xlabel('Selling price of Hydrogen [$/kg]')
plt.ylabel('Revenue/M$')
plt.legend()
plt.show()
```
